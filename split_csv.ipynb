```python
import pandas as pd
df = pd.read_csv("southeast/southeast.csv")
df = df.drop('index', axis=1)
df.columns = ['date', 'hour', 'prcp', 'stp', 'smax', 'smin', 'gbrd', 'temp', 'dewp', 'tmax', 'tmin', 'dmax', 'dmin', 'hmax', 'hmin', 'hmdy', 'wdct', 'gust', 'wdsp', 'region', 'state', 'station', 'station_code', 'latitude', 'longitude', 'height']
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year
df['year'].nunique()
```

```python
df['year'] = df['year'].astype(str)
df.dtypes
```

```python
for name, group in df.groupby('year'):
    group.to_csv('smallerCSV/' + '{}.csv'.format(name), index=False)
```
